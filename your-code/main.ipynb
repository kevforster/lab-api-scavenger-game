## Challenge 3: Hidden Cold Joke

Using Python, call Github API to find out the cold joke contained in the 24 secret files in the following repo:

https://github.com/ironhack-datalabs/scavenger

The filenames of the secret files contain .scavengerhunt and they are scattered in different directories of this repo. The secret files are named from .0001.scavengerhunt to .0024.scavengerhunt. They are scattered randomly throughout this repo. You need to search for these files by calling the Github API, not searching the local files on your computer.

#### 1. Find the secret files.

`````python
# your code here
username='kevforster'
token= 'f0df14f5463c11b74d35c56ff3f6a77d6589e53a'
url="https://api.github.com/repos/ironhack-datalabs/scavenger/contents"
response=requests.get (url, auth=(username,token))
response
`````

`````python
result=response.json()

result
`````

`````python
filenames_sh=[]
for i in range(1,len(result)):
    url=result[i]['url']
    response=requests.get (url, auth=(username,token))
    files=response.json()
    for k in files:
            filenames=k['name']
            if filenames.endswith('scavengerhunt'):
                filenames_sh.append(filenames)
print(filenames_sh)
        
`````

#### 2.  Sort the filenames ascendingly.

`````python
# your code here
serie_filename=pd.Series(filenames_sh)
serie_filename
`````

`````python
serie_filename.sort_values()
`````

#### 3. Read the content of each secret files into an array of strings.
Since the response is encoded, you will need to send the following information in the header of your request:
````python
headers = {'Accept': 'application/vnd.github.v3.raw'}
````

`````python
username='kevforster'
token= 'f0df14f5463c11b74d35c56ff3f6a77d6589e53a'
url="https://api.github.com/repos/ironhack-datalabs/scavenger/contents"
headers = {'Accept': 'application/vnd.github.v3.raw'}
response=requests.get (url, auth=(username,token), headers=headers)
response
result=response.json()
result[1]
`````

`````python
content_sf=[]
for i in range(1,len(result)):
    url=result[i]['url']
    response=requests.get (url, auth=(username,token),headers=headers)
    files=response.json()
    for k in files:
            filenames=k['name']
            if filenames.endswith('scavengerhunt'):
                sf_url=k['url']
                response2=requests.get (sf_url, auth=(username,token),headers=headers)
                content=response2.text
                content_sf.append(content)

print(content_sf)
`````

`````python
serie_content=pd.Series(content_sf)
serie_content
`````

`````python
serie_concat=pd.concat([serie_filename, serie_content], axis=1)
serie_concat.columns=['name','secret']
serie_concat=serie_concat.sort_values(by='name', ascending=True )
`````

`````python
serie_concat = serie_concat.replace('\n','', regex=True)
`````

`````python
serie_concat
`````

#### 4. Concatenate the strings in the array separating each two with a whitespace.
